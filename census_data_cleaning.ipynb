```python
import pandas as pd
```

```python
# Import ny income file
income_df = pd.read_csv('https://raw.githubusercontent.com/tawney-kirkland/Metis_Project_1/main/data/ny_income.csv',
                       usecols = ['GeoType', 'GeogName', 'GeoID', 'Borough','Pop16plE','HH2E',
                              'HHI75t99E','HI100t149E','HI150t199E', 'HHI200plE', 'MdHHIncE', 'MnHHIncE'])
```

```python
income_df.rename(columns = {'Pop16plE': 'above_16_pop',
                            'HH2E': 'total_households',
                           'HHI75t99E': 'income_75_99k',
                           'HI100t149E': 'income_100_149k',
                            'HI150t199E': 'income_150_199k',
                            'HHI200plE': 'income_200k_over',
                           'MdHHIncE': 'income_median',
                           'MnHHIncE': 'income_mean'},
                 inplace=True
                )
```

```python
# Calculate the percent of households in each area within the specified income ranges
income_df['percent_75_99k'] = income_df['income_75_99k'] / income_df['total_households'] 
income_df['percent_100_149k'] = income_df['income_100_149k'] / income_df['total_households'] 
income_df['percent_150_199k'] = income_df['income_150_199k'] / income_df['total_households']
income_df['percent_200_over'] = income_df['income_200k_over'] / income_df['total_households']

# Sum percent of households with more than 100k income to determine a thresh hold
income_df['percent_above_100k'] = income_df['percent_100_149k'] + income_df['percent_150_199k'] + income_df['percent_200_over']
```
